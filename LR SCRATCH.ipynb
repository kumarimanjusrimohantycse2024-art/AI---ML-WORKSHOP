```python
import numpy as np

class SimpleLinearRegression:
    def __init__(self):
        self.m = None  # Slope
        self.c = None  # Intercept

    # Fit the model to the data using the formula for linear regression
    def fit(self, x, y):
        # Convert x and y to numpy arrays for easier manipulation
        x = np.array(x)
        y = np.array(y)
        
        # Calculate the means of x and y
        x_mean = np.mean(x)
        y_mean = np.mean(y)
        
        # Calculate the slope (m)
        numerator = np.sum((x - x_mean) * (y - y_mean))
        denominator = np.sum((x - x_mean) ** 2)
        self.m = numerator / denominator
        
        # Calculate the intercept (c)
        self.c = y_mean - self.m * x_mean
    
    # Predict the y values for the given x values using the learned model
    def predict(self, x):
        x = np.array(x)
        return self.m * x + self.c
    
    # Calculate Mean Squared Error (MSE)
    def mse(self, x, y):
        y_pred = self.predict(x)
        mse_value = np.mean((y - y_pred) ** 2)
        return mse_value
    
    # Calculate R-squared (Coefficient of Determination)
    def r2_score(self, x, y):
        y_pred = self.predict(x)
        
        # Total sum of squares (variance of the y values)
        total_sum_of_squares = np.sum((y - np.mean(y)) ** 2)
        
        # Residual sum of squares (error in the predictions)
        residual_sum_of_squares = np.sum((y - y_pred) ** 2)
        
        # R-squared = 1 - (residual sum of squares / total sum of squares)
        r2 = 1 - (residual_sum_of_squares / total_sum_of_squares)
        return r2

# Example Usage:

# Sample data for testing
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# Create the model
model = SimpleLinearRegression()

# Fit the model to the data
model.fit(x, y)

# Make predictions
y_pred = model.predict(x)

# Calculate Mean Squared Error
mse_value = model.mse(x, y)

# Calculate R-squared
r2_value = model.r2_score(x, y)

print(f"Slope (m): {model.m}")
print(f"Intercept (c): {model.c}")
print(f"Predictions: {y_pred}")
print(f"Mean Squared Error: {mse_value}")
print(f"R-squared: {r2_value}")
```

```python
class MultipleLinearRegression:
    def __init__(self):
        self.betas = []
        
    def concat_ones(self,x):
        ones = np.ones((x.shape[0], 1))
        return np.concatenation([ones, x], axis = 1)
    
    def fit(self, x, y):
        x = self.concat_ones(x)
        xt = x.T
        xtx = np.dot(xt, x)
        xtxinv = np.linalg.inv(xtx)    
        xty = np.dot(xt, y)
        self.betas = np.dot(xtxinv, xty)
        
    def predict(self, x):
        x = self.concat_ones(x)
        return np.dot(x, self.betas)
    
    def mse(self, x, y):
        return np.mean((y - self.predict(x))**2)
    
    def r2_score(self, x, y):
        rss = np.sum((y - self.predict(x))**2)
        tss = np.sum((y - np.mean(y)) ** 2)
        return 1 - (rss / tss)
    
        
        
    
```
